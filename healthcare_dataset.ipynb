```python
import pandas as pd
df= pd.read_csv('Dataset Kesehatan.xls - healthcare_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df['Billing Amount'] = df['Billing Amount'].astype(str).str.replace(',','.', regex= False).astype(float)
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])
```

```python
Kategori_cols = ['Age', 'Gender', 'Blood Type', 'Medical Condition', 'Insurance Provider', 'Test Results']
for col in Kategori_cols:
  df[col] = df[col].astype('category')
```

```python
df.isnull().sum()[df.isnull().sum() > 0]
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df['Billing Amount'].describe()
```

```python
df= df[df['Billing Amount']>= 0]
```

```python
Q1 = df['Billing Amount'].quantile(0.25)
Q3 = df['Billing Amount'].quantile(0.75)

IQR = Q3 - Q1
lower  = Q1 - 1.5 *IQR
Upper = Q3 + 1.5 * IQR
```

```python
# Metode B: Ganti Outlier dengan Median
Median_BA = df['Billing Amount'].median()

# Ganti nilai outlier dengan median
df.loc[(df['Billing Amount'] < lower) | (df['Billing Amount'] > Upper), 'Billing Amount'] = Median_BA
```

```python
df['Billing Amount'].describe()
```

```python
df['Name'] = df['Name'].str.title()
df.head()
```
